```python
from __future__ import print_function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from fbprophet import Prophet

%matplotlib inline
```

```python
# Data obtained from https://finance.yahoo.com/quote/%5EGSPC/history/

data = pd.read_csv("DowJones2008-2018.csv")
```

```python
data.dtypes
```

```python
data['Datetime'] = pd.to_datetime(data['Date'])
```

```python
data.tail(5)
```

```python
# Declare dataframe with column 'ds' and 'y' as required to feed to Prophet. 
# Carry out log transform on y

df = data[['Datetime','Close']][data.Datetime <= pd.datetime(2017,6,1)]
df.columns = ['ds', 'y']
df['y'] = np.log(df['y'])
m = Prophet(weekly_seasonality = False)
m.fit(df);
```

```python
future = m.make_future_dataframe(periods = 365, freq = 'D', include_history=True)
forecast = m.predict(future)
```

```python
m.plot(forecast,ylabel = 'Log(Weighted_Price)',xlabel='Date');
```

```python
m.plot_components(forecast);
```

```python
def plot_actual(forecast, data, start, end):

    dummy = forecast[['ds','yhat','yhat_lower','yhat_upper']].set_index('ds')
    dummy['log_yhat'] = dummy['yhat']
    dummy['yhat'] = np.exp(dummy['yhat'])
    dummy['yhat_lower'] = np.exp(dummy['yhat_lower'])
    dummy['yhat_upper'] = np.exp(dummy['yhat_upper'])

    plt.figure(figsize = (10,6))

    d = dummy.loc[start:end]
    d = d.join(data[['Datetime','Close']].set_index('Datetime'))
    d['yhat'].plot()
    d['Close'].plot(marker = '.', c = 'black')
    plt.fill_between(d.index, d['yhat_lower'], d['yhat_upper'],color='#0072B2', alpha=0.2)
    plt.grid();
```

```python
start = pd.datetime(2015,1,1)
end = pd.datetime(2017,12,1)

plot_actual(forecast,data,start, end)
```
